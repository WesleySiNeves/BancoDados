## query que retorna a quantidade de memoria do SQL por query

```sql
SELECT er.session_id,
       SUBSTRING(   st.text,
                    (er.statement_start_offset / 2) + 1,
                    ((CASE er.statement_end_offset
                          WHEN -1 THEN
                              DATALENGTH(st.text)
                          ELSE
                              er.statement_end_offset
                      END - er.statement_start_offset
                     ) / 2
                    ) + 1
                ) AS statement_text,
       mg.dop,                                               --grau de paralelismo da consulta,
       er.wait_type,
       er.wait_time,
       mg.requested_memory_kb / 1024 AS requested_memory_MB, --Quantidade total solicitada de memória,
       mg.granted_memory_kb / 1024 AS granted_memory_MB,     --Total de memória realmente concedido em quilobytes. Poderá ser NULL se a memória ainda não tiver sido concedida
       mg.used_memory_kb / 1024 AS used_memory_kb_MB,        --Memória física usada neste momento em quilobytes.
       mg.query_cost,                                        --Custo de consulta estimado
       mg.timeout_sec,                                       --Tempo limite em segundos antes de esta consulta desistir da solicitação de concessão de memória.
       mg.queue_id,                                          --ID da fila de espera em que esta consulta aguarda concessões de memória. NULL se a memória já tiver sido,
       mg.wait_order,                                        --Ordem sequencial das consultas de espera dentro do especificado queue_id,
       mg.is_next_candidate,                                 --Candidato para a próxima concessão de memória.
       CONVERT(XML, qp.query_plan) AS query_plan
FROM sys.dm_exec_requests er
    CROSS APPLY sys.dm_exec_sql_text(er.sql_handle) st
    CROSS APPLY sys.dm_exec_text_query_plan(er.plan_handle, er.statement_start_offset, er.statement_end_offset) qp
    INNER JOIN sys.dm_exec_query_memory_grants mg
        ON er.session_id = mg.session_id
ORDER BY mg.requested_memory_kb DESC;

```
