```sql
$mensagem = "(a) consultar saldo, (b) saque, (c) depsito  e (d) sair"


$opcao = [string](Read-Host $mensagem)

$saldoBancario = 0;
$valorSaque = 0;
$valorDeposito = 0;



while (! ($opcao  -eq 'd')) {
 
    switch ($opcao) {
        'a' { 
                 Write-Host 'Seu saldo atual  de:',$saldoBancario;
                $opcao = [string](Read-Host $mensagem)
                break;
            }
        'b' {
                $valorSaque = [float](Read-Host 'Digite o valor do saque:');

                if ($valorSaque -gt $saldoBancario) {
                    Write-Host 'Saldo insuficiente.! seu saldo atual  de :',$saldoBancario;
                    $opcao = [string](Read-Host $mensagem)
                    break;
                }
                else {
                    $saldoBancario = $saldoBancario -  $valorSaque;
                    Write-Host 'Saque efetuado com sucesso! seu saldo atual  de :',$saldoBancario;
                    $opcao = [string](Read-Host $mensagem)
                    break;
                }
            }
        'c' {
                $valorDeposito =[float](Read-Host 'Digite o valor do deposito:');
                $saldoBancario = $saldoBancario +  $valorDeposito;
                Write-Host 'Deposito efetuado com sucesso! seu saldo atual  de :',$saldoBancario;
                $opcao = [string](Read-Host $mensagem)
                
            }
        Default { Write-Host 'Opo invalida'}
    }
}

```
