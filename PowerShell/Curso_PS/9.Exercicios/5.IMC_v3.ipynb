## Receba o nome , idade , peso e altura e calcule o IMC

```powershell
$nome= Read-Host 'Digite seu nome:'

$idade = Read-Host  'Digite sua idade:'

$peso = [float](Read-Host  'Digite seu peso:')

$altura = [float](Read-Host  'Digite sua altura:')

$imc = $peso / ($altura * $altura)

$mensagem = ""

if($imc -lt 15) { $mensagem  = "Extremamente abaixo do peso" }
if($imc -ge 15 -and   $imc -lt 16) { $mensagem  = "Gravemente abaixo do peso" }   
if($imc -ge 16 -and   $imc -lt 18.5) { $mensagem  = "Abaixo do pesso Ideal" }   
if($imc -ge 18.5 -and   $imc -lt 25) { $mensagem  = "Faixa de peso ideal" }   
if($imc -ge 25 -and   $imc -lt 30) { $mensagem  = "Sobrepeso" }   
if($imc -ge 30 -and   $imc -lt 35) { $mensagem  = "Obesidade grau I" }   
if($imc -ge 35 -and   $imc -lt 40) { $mensagem  = "Obesidade grau II (grave)" }   
if($imc -ge 40) { $mensagem  = "Obesidade grau III (morbida)" }   
Write-Host $imc
```
