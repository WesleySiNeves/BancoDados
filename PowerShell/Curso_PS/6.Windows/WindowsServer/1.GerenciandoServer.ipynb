## Para  gerenciar o windows server primeiro devemos importar um modulo do power shell

```powershell
Import-Module ServerManager
```

\## Busca todos os itens instalados

```powershell
Get-WindowsFeature
```

\## Comando para instalar o IIS no servidor

```powershell
Install-WindowsFeature -Name Web-Server -IncludeManagementTools
```

\## Verificar se algo já fo instalado no servidor -Começa com

```powershell
Get-WindowsFeature -Name Web-*
```
