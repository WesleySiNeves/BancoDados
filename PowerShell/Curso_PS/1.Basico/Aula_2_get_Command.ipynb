Recuperar todos os comandos de um modulo especifico

```powershell
Get-Command -Module Az.Accounts
```

## Recuperar todos os comandos

```powershell
Get-Command
```

## especificando comandos

```powershell
Get-Command Get-Host
```

## Todos os comandos que terminen com host

```powershell
Get-Command *-host
```

## Comandos que tenha 3 letras e termine com Host

```powershell
Get-Command ???-host
```
